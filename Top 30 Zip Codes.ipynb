```python
import pandas as pd 

#Import csv files generated by each group member
file_2015 = "2015_Top30zipcodes.csv"
file_2016 = "2016_Top30zipcodes.csv"
file_2017 = "2017_Top30codes"
file_2018 = "2018_Top30zipcodes.csv"

#read each csv file
df1 = pd.read_csv(file_2015)
df2 = pd.read_csv(file_2016)
df3 = pd.read_csv(file_2017)
df4 = pd.read_csv(file_2018)

#concatenate all of the csv files into one csv file
merge = pd.concat([df1["Zip Code"],df2["Zip Code"],df3["Zip Code"],df4["zipcode"]],ignore_index=True)
#merge and convert to dataframe
convert = merge.to_frame()

#create final df and con
final_df = convert.rename(index=str, columns={0: "Zip Codes"})
final_df.to_csv("consolidated.csv", sep=",")
zipcodes = final_df.groupby(["Zip Codes"])["Zip Codes"].count()
zipcodes
```
