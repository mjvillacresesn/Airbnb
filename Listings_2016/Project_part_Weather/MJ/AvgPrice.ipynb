```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#added -files all yrs
file_input_15 = './topzips_raw_2015.csv'
file_input_16 = './topzips_raw_2016.csv'
file_input_17 = './topzips_raw_2017.csv'
file_input_18 = './topzips_raw_2018.csv'
```

```python
#2015 : May - Dec, (Missing July Data)
file_15 = pd.read_csv(file_input_15)
file_15
#avg price / month (2016) for oveall (2015-2018) "top 30" zipcodes
by_month_15 = file_15.groupby('year_month').mean()
by_count_15 = file_15.groupby('year_month').count()
#print(by_count_15)
#new_15 = by_month_15
b_15 = by_month_15["price"]
c_15 = by_count_15["zipcode"]

#2016 : Jan - Dec, (Missing March Data)
file_16 = pd.read_csv(file_input_16)
#avg price / month (2016) for oveall (2015-2018) "top 30" zipcodes
by_month_16 = file_16.groupby('year_month').mean()
by_count_16 = file_16.groupby('year_month').count()
#print(by_count_16)
#new = by_month
b = by_month_16["price"]
c_16 = by_count_15["zipcode"]

#2017 :Jan - Dec
file_17 = pd.read_csv(file_input_17)
#avg price / month (2016) for oveall (2015-2018) "top 30" zipcodes
by_month_17 = file_17.groupby('year_month').mean()
by_count_17 = file_17.groupby('year_month').count()
#new = by_month
b_17 = by_month_17["price"]
c_17 = by_count_15["zipcode"]

#2018 : Jan - Dec
file_18 = pd.read_csv(file_input_18)
#avg price / month (2016) for oveall (2015-2018) "top 30" zipcodes
by_month_18 = file_18.groupby('Date').mean()
by_count_18 = file_18.groupby('Date').count()
#new = by_month
b_18 = by_month_18["price"]
c_18 = by_count_15["zipcode"]
#print(b_18)


# y-axis == by_month["price"]
# x-axis == by_month["year_month"]
# s == number of listing / zip code
```

```python
#Avg Mean / Month 
concat1 = pd.concat([b_15, b, b_17, b_18], axis=0)
#print(concat1)
# Number of Lsitings / per "top 30" zipcodes
concat2 = pd.concat([c_15, c_16, c_17, c_18], axis=0)
#print(concat2)
```
