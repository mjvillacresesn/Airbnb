```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Import 2016 CSV
df_15 = pd.read_csv("2015_Top30zipcodes.csv")

#Remove random column
df_2015 = df_15[['Zip Code','Mean', 'Median']]
df_2015
```

```python
#Import 2016 CSV
df_16 = pd.read_csv("2016_Top30zipcodes.csv")

#Remove random column
df_2016 = df_16[['Zip Code','Mean', 'Median']]
df_2016
```

```python
#Merge 2015 and 2016
df_1516 = pd.merge(df_2015, df_2016, on="Zip Code", how="outer", suffixes=("_2015", "_2016"))
df_1516
```

```python
#Import 2017 CSV
df_2017 = pd.read_csv('2017_Top30codes')
df_2017
```

```python
#Merge 2015/2016 with 2017
df_151617 = pd.merge(df_1516, df_2017, on="Zip Code", how="outer")
df_151617
#df_151617.rename(columns={'Mean': 'Mean_2017', 'Median': 'Median_2017'})
```

```python
#Import 2018 CSV
df_18 = pd.read_csv('2018_Top30zipcodes_edited.csv')

#Remove random column
df_2018 = df_18[['Zip Code','Mean', 'Median']]
df_2018

df_2018
```

```python
#Merge 2015/2016/2017 with 2018
df_15161718 = pd.merge(df_151617, df_2018, on="Zip Code", how="outer", suffixes=("_2017","_2018"))
df_15161718
```

```python
df_15161718.to_csv('Merged_Top30codes.csv', index=False)
```
